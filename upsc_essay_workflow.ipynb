```python
from langgraph.graph import StateGraph, START, END
from langchain_openai import ChatOpenAI
from dotenv import load_dotenv
from typing import TypedDict
from pydantic import BaseModel, Field
```

```python
load_dotenv()
```

```python
model = ChatOpenAI(model="gpt-4o-mini")
```

```python
class EvaluationSchema(BaseModel):
    feedback: str = Field(description="Detailed feedback for the essay")
    score: int = Field(description="Score for the essay, between 0 and 10", ge=0, le=10)
```

```python
structured_model = model.with_structured_output(EvaluationSchema)
```

```python
# Write a random essay of 100 words
essay = """
The rapid advancement of technology has transformed the way we live, work, and communicate. From smartphones to artificial intelligence, technology has become an integral part of our daily lives. It has revolutionized industries, improved healthcare, and connected people across the globe. However, it also poses challenges such as privacy concerns and job displacement. As we continue to innovate, it is crucial to strike a balance between embracing technological progress and addressing its potential drawbacks. By fostering responsible innovation and ethical considerations, we can harness the benefits of technology while mitigating its risks.
"""
```

```python
prompt = f'Evaluate the language quality of the following essay and provide feedback and a score between 0 and 10:\n\n{essay}'
structured_model.invoke(prompt).feedback
```
